```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("dataset_market_basket.csv")
```

```python
df
```

```python
df.sum()
```

```python
# 2 Item set
from itertools import permutations

#membuat perbandingan data
rules = list(permutations(df.columns, 2))
rules
```

```python
# mengfilter data yang dimana hanya item yang ada saja yang akan terhitung

rule_baru = []
for rule in rules:
  if len(df[(df[rule[0]]==1) & (df[rule[1]]==1)]) >0:
    rule_baru.append(rule)
rule_baru
```

```python
# melihat jumlah data 2 item set
hitung = {} #Ganti dictionary
for rule in rule_baru:
  banyak_data= len(df[
      (df[rule[0]]==1) & (df[rule[1]]==1)
      ])
  hitung[str(rule)] = banyak_data

hitung
```

```python
# List support
supp = [i/len(df)*100 for i in hitung.values()]
print(supp)
print("---")
for item,support in hitung.items():
  print(f"{item} support = {round(support/len(df)*100)}%")
```

```python
# List Confidence

print("---")
for item,ab in hitung.items():
  i = eval(item)
  print(f"{i[0]}->{i[1]} Confidence = {ab/df[i[0]].sum()*100}%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# membaca dataset dan menyimpannya dalam variabel df
df = pd.read_csv('insurance.csv')

# visualisasi distribusi usia
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='age', kde=True, bins=20)
plt.title('Distribusi Usia Klien Asuransi Kesehatan')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.show()

# visualisasi hubungan antara usia dan biaya klaim
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='age', y='charges', hue='smoker')
plt.title('Hubungan Antara Usia dan Biaya Klaim Asuransi Kesehatan')
plt.xlabel('Usia')
plt.ylabel('Biaya Klaim')
plt.show()

# visualisasi perbedaan biaya klaim antara perokok dan bukan perokok
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='smoker', y='charges')
plt.title('Perbedaan Biaya Klaim Asuransi Kesehatan antara Perokok dan Bukan Perokok')
plt.xlabel('Perokok')
plt.ylabel('Biaya Klaim')
plt.show()

# visualisasi hubungan antara indeks massa tubuh dan biaya klaim
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='bmi', y='charges')
plt.title('Hubungan Antara Indeks Massa Tubuh dan Biaya Klaim Asuransi Kesehatan')
plt.xlabel('Indeks Massa Tubuh (BMI)')
plt.ylabel('Biaya Klaim')
plt.show()
```
